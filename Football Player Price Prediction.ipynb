**Normalizing the data**

```python
A = d['fpl_value'].values
B = d['market_value'].values
A_scaler = MinMaxScaler()
A = A_scaler.fit_transform(A.reshape(-1,1))
A = A[ : , -1]
B_scaler = MinMaxScaler()
B = B_scaler.fit_transform(B.reshape(-1,1))
B = B[ : , -1]
```

```python
Atrain, Atest, Btrain, Btest = train_test_split(A, B, test_size=0.2)
```

**LINEAR REGRESSION**

```python
def error(m, x, c, t):  #ERROR FUNC
    N = x.size
    e = sum(((m * x + c) - t)**2)
    return e * 1/(2 * N)
```

```python
def update(m, x, c, t, learning_rate):  # UPDATE FUNC
    grad_m = sum(2*((m*x+c)-t)*x)
    grad_c = sum(2*((m*x+c)-t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m,c
```

```python
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = []
    mc_values = []
    for i in range(iterations):
        e = error(m,x,c,t)
        if e < error_threshold:
            print("Error less than the threshold. Stopping the gradient descent.")
            break
        error_values.append(e)
        m,c = update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m, c, error_values, mc_values
```

```python
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001
m, c, error_values, mc_values = gradient_descent(init_m, init_c, Atrain, Btrain, learning_rate, iterations, error_threshold)
```
